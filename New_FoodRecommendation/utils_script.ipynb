```python
import pandas as pd

# Input and output file paths
input_file = "./south_indian_food.csv"
output_file = "./south_indian_food_with_id.csv"

# Load dataset
df = pd.read_csv(input_file)

# Create food_id column (1,2,3,...)
df.insert(0, "food_id", range(1, len(df) + 1))

# Save updated file
df.to_csv(output_file, index=False)

print("food_id column added successfully.")
print(f"Saved file: {output_file}")
```
